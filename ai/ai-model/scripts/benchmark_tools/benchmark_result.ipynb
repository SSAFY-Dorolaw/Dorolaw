```python
import json
```

```python
import os
```

```python
import matplotlib.pyplot as plt
```

```python
# 디렉토리 및 모델 리스트
result_dir = "../../work_dirs/work_dirs/benchmark_result"
model_names = ["TSN", "SlowFast", "Swin"]

accuracies = {}

for model in model_names:
    json_path = os.path.join(result_dir, f"{model}_result.json")
    with open(json_path, "r") as f:
        data = json.load(f)

    total = len(data)
    correct = sum(d["pred_label"] == d["gt_label"] for d in data)
    acc = correct / total
    accuracies[model] = acc

    print(f"{model} Accuracy: {acc:.2%} ({correct}/{total})")

# 시각화
plt.figure(figsize=(8, 5))
plt.bar(accuracies.keys(), accuracies.values())
plt.ylim(0, 1)
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.grid(axis='y')
plt.show()
```
